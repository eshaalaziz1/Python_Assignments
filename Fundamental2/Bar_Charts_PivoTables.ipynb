## Barchart using plt.bar()

```python
from unicodedata import category
import pandas as pd
import matplotlib.pyplot as plt

categories = ["Category A", "Category B", "Category C", "Category D"]
values = [23, 45, 56, 78]

# creating the bar chart
plt.figure(figsize=(8,5))
plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar chart example")
plt.show()



```

## Barchart using Dataframe's plot() Method

```python
import pandas as pd

df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'business': [500, 400, 300, 500, 1000]

})

df.set_index('city')['business'].plot(kind='bar', figsize=(10, 6), color='green', edgecolor='black', title='Number of Business Per City')
plt.xticks(rotation=45)
plt.ylabel('Number of Business')
plt.show()
```

## Plotting a Pivot Table

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'category_0': ['Bars', 'Restaurants', 'Cafes', 'Fastfood'],
    'stars': [4.2, 3.8, 4.5, 2.8]

})

# filtering data for categories
cat = df['category_0'].isin(['Bars', 'Restaurants', 'Cafes', 'Fastfood'])
df_cat = df[cat]

#Creating a pivot table to get average star ratings by category
pivot_state_cat = pd.pivot_table(df_cat, index=['category_0'], values='stars', aggfunc='mean')

pivot_state_cat.plot(kind='bar', figsize=(10, 6),
color='purple', title='Average Star Rating for Food Places')
plt.xlabel('Categories')
plt.ylabel('Average Star Rating')
plt.show()
```
