# 1. Scatter Plot: Price vs. Year (with trendline)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset into a Pandas DataFrame
df = pd.read_csv("car_price_dataset.csv")

# 1. Scatter Plot: Price vs. Year (with trendline)
plt.figure(figsize=(8, 5))  # Set figure size

# Create a scatter plot of car prices based on production year
plt.scatter(df["Year"], df["Price"], alpha=0.5, color="blue", label="Car Prices")

# Calculate a linear trendline (y = mx + b) using numpy's polyfit
m, b = np.polyfit(df["Year"], df["Price"], 1)

# Plot the trendline in red
plt.plot(df["Year"], m * df["Year"] + b, color="red", label="Trendline")

# Set axis labels and title
plt.xlabel("Year")
plt.ylabel("Price")
plt.title("Car Price vs. Year")

# Add a legend to differentiate scatter points and the trendline
plt.legend()

# Display the plot
plt.show()
```

# 2. Box Plot: Price distribution by Fuel Type

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 2. Box Plot: Price distribution by Fuel Type
plt.figure(figsize=(8, 5))  # Set figure size

# Generate a box plot to visualize the price distribution for each fuel type
df.boxplot(column="Price", by="Fuel_Type", vert=False, grid=False)

# Set axis labels and title
plt.xlabel("Price")
plt.ylabel("Fuel Type")
plt.title("Price Distribution by Fuel Type")

# Remove the default automatic title from the Pandas boxplot
plt.suptitle("")

# Display the plot
plt.show()
```

# 3. Histogram + Normal Approximation: Mileage distribution

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 3. Histogram + Normal Approximation: Mileage distribution
plt.figure(figsize=(8, 5))  # Set figure size

# Create a histogram to visualize the mileage distribution
plt.hist(df["Mileage"], bins=30, color="green", alpha=0.6, density=True, label="Mileage Distribution")

# Compute the mean and standard deviation of mileage
mileage_mean = np.mean(df["Mileage"])
mileage_std = np.std(df["Mileage"])

# Generate a normal distribution curve using the computed mean and std deviation
x = np.linspace(df["Mileage"].min(), df["Mileage"].max(), 100)
y = (1 / (mileage_std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mileage_mean) / mileage_std) ** 2)

# Plot the normal approximation curve in red
plt.plot(x, y, color="red", label="Normal Approximation")

# Set axis labels and title
plt.xlabel("Mileage (km)")
plt.ylabel("Density")
plt.title("Mileage Distribution with Approximation")

# Add a legend to differentiate histogram and approximation curve
plt.legend()

# Display the plot
plt.show()
```

# 4. Bar Chart: Top 10 Car Brands by Average Price

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# 4. Bar Chart: Top 10 Car Brands by Average Price
# Compute the average price per brand and select the top 10 most expensive brands
brand_avg_price = df.groupby("Brand")["Price"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(8, 5))  # Set figure size

# Create a bar chart to visualize the top 10 most expensive car brands (on average)
plt.bar(brand_avg_price.index, brand_avg_price.values, color="purple")

# Set axis labels and title
plt.xlabel("Car Brand")
plt.ylabel("Average Price")
plt.title("Top 10 Car Brands by Average Price")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

# 5. Heatmap: Correlation Between Numerical Features

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 5. Heatmap: Correlation Between Numerical Features
# Select only numerical columns for correlation analysis
numerical_cols = ["Year", "Engine_Size", "Mileage", "Owner_Count", "Price"]
correlation_matrix = df[numerical_cols].corr()  # Compute correlation matrix

plt.figure(figsize=(8, 5))  # Set figure size

# Create a heatmap visualization using imshow
plt.imshow(correlation_matrix, cmap="coolwarm", interpolation="none")

# Add a color bar to indicate correlation strength
plt.colorbar()

# Set x and y tick labels using column names
plt.xticks(range(len(numerical_cols)), numerical_cols, rotation=45)
plt.yticks(range(len(numerical_cols)), numerical_cols)

# Set the title for the heatmap
plt.title("Feature Correlation Heatmap")

# Display the plot
plt.show()
```
